```python
# pulling the number of dates
data = engine.execute("SELECT tot_death AS Deaths FROM usa_data  WHERE state='ND' Order By submission_date Desc LIMIT 1;")
for record in data:
    print(record)
```

```python
# verify the data ... it is the 8th piece of data

data = engine.execute("SELECT tot_cases, state FROM (SELECT tot_cases, state, ROW_NUMBER() OVER(PARTITION BY state ORDER BY submission_date DESC) as row_num FROM usa_data) table1 WHERE row_num = 1;")
for record in data:
    print(record)
```
